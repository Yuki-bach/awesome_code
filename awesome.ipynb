### 3章　集約

```python
result = researve_tb \
  .groupby('hotel_id') \
  .agg({'reserve_id': 'count', 'customer_id': 'nunique'}) 
result.reset_index(inplace=True)
result.columns = ['hotel_id', 'rsv_cnt', 'cus_cnt']
result.head()
```

```python
result = researve_tb \
  .groupby(['hotel_id', 'people_num'])['total_price'] \
    .sum().reset_index()
result.rename(columns = {'total_price': 'price_sum'}, inplace=True)
result.head()
```

```python
result = researve_tb \
  .groupby('hotel_id')['total_price'] \
    .agg(['max', 'min', 'mean', 'median', lambda x: np.percentile(x, q=20)]) \
      .reset_index()
result.columns = ['hotel_id', 'price_max', 'price_min', 'price_mean', 'price_median', 'price_20per']
result.head()
```

```python
result = researve_tb \
  .groupby('hotel_id')['total_price'] \
    .agg(['var', 'std']) \
      .reset_index()
result.columns = ['hotel_id', 'price_var', 'price_std']
result.fillna(0, inplace=True)
result.head()      
```

```python
# 予約日時の順位を古い順につける
researve_tb['researve_datetime'] = pd.to_datetime(
  researve_tb['reserve_datetime'], format='%Y-%m-%d %H:%M:%S')

researve_tb['log_no'] = researve_tb \
  .groupby('customer_id')['researve_datetime'] \
    .rank(method='first', ascending=True)
researve_tb.head()
```

```python
# 予約数の順位をつける
rsv_cnt_tb = researve_tb \
  .groupby('hotel_id') \
    .size().reset_index()
rsv_cnt_tb.columns = ['hotel_id', 'rsv_cnt']    
rsv_cnt_tb['rsv_cnt_rank'] = rsv_cnt_tb['rsv_cnt'] \
  .rank(ascending=False, method='min')
rsv_cnt_tb.drop('rsv_cnt', axis=1, inplace=True)
rsv_cnt_tb.head()  
```

### 4章　結合

```python
hotel_tb = pd.read_csv('awesomebook-master/data/hotel.csv', encoding='utf-8')
hotel_tb.head()
```

```python
pd.merge(researve_tb.query('people_num == 1'), hotel_tb.query('is_business'),
on='hotel_id', how='inner').head()
```

```python
# 全ホテルに対して、レコメンド候補を作成する
import gc

# 各エリアのホテル数を集計
small_area_mst = hotel_tb \
  .groupby(['big_area_name', 'small_area_name']) \
  .size().reset_index()
small_area_mst.columns = ['big_area_name', 'small_area_name', 'hotel_cnt']

# ホテル数が20件以上のエリアは、小エリア名を、それ以外は大エリア名を結合キーとする
# small_area_mst[small_area_name, join_area_id]
small_area_mst['join_area_id'] = \
  np.where(small_area_mst['hotel_cnt']-1 >= 20, small_area_mst['small_area_name'], small_area_mst['big_area_name'])
small_area_mst.drop(['hotel_cnt', 'big_area_name'], axis=1, inplace=True)

# ホテルマスタと結合し、結合キーを削除
base_hotel_mst = pd.merge(hotel_tb, small_area_mst, on='small_area_name') \
  .loc[:, ['hotel_id', 'join_area_id']]

del small_area_mst
gc.collect()

# recommend_hotel_mst[join_area_id, rec_hotel_id]
recommend_hotel_mst = pd.concat([
  hotel_tb[['small_area_name', 'hotel_id']] \
    .rename(columns={'smal_area_name': 'join_area_id'}, inplace=False),
  hotel_tb[['big_area_name', 'hotel_id']] \
    .rename(columns={'big_area_name': 'join_area_id'}, inplace=False)
])
recommend_hotel_mst.rename(columns={'hotel_id': 'rec_hotel_id'}, inplace=True)

# 結合
pd.merge(base_hotel_mst, recommend_hotel_mst, on='join_area_id') \
  .loc[:, ['hotel_id', 'rec_hotel_id']] \
    .query('hotel_id != rec_hotel_id')
base_hotel_mst.head()    
```

```python
# n件前のレコードを取得する
result = researve_tb \
  .groupby('customer_id') \
    .apply(lambda group: group.sort_values(by='researve_datetime', axis=0, inplace=False))

result['before_price'] = \
  pd.Series(result['total_price'].shift(periods=2))
result.head()      
```

```python
# 対数化：値が大きくなるほど、値の差の意味を小さくしたい時に使う
researve_tb['total_price_log'] = \
  researve_tb['total_price'].apply(lambda x: np.log(x/1000 + 1))
researve_tb.head()  
```

```python
# 数値のカテゴリ化
customer_tb = pd.read_csv('awesomebook-master/data/customer.csv', encoding='utf-8')
customer_tb['age_rank'] = (np.floor(customer_tb['age'] / 10) * 10).astype('category')
customer_tb.head()
```

```python
# 正規化
# 過学習を防ぐために、データの値を0～1の範囲に収める
# 1. 平均0、分散1に変換。(入力値-入力値の平均/)入力値の標準偏差
# 2. 最大値1、最小値0に変換
from sklearn.preprocessing import StandardScaler

researve_tb['people_num'] = researve_tb['people_num'].astype(float)

ss = StandardScaler()
result = ss.fit_transform(researve_tb[['people_num', 'total_price']])

researve_tb['people_num_normalized'] = [x[0] for x in result]
researve_tb['total_price_normalized'] = [x[1] for x in result]
```

```python
# 外れ値の除去
# 平均値から標準偏差の3倍以内の値に収まるレコードに絞る
researve_tb = researve_tb[
  (abs(researve_tb['total_price'] - np.mean(researve_tb['total_price'])) /
  np.std(researve_tb['total_price']) <= 3)
  ].reset_index()
researve_tb.head()
```

```python
# 数値の補完
# 多重代入法や最尤法が一般的
production_missing_num = pd.read_csv('awesomebook-master/data/production_missing_num.csv', encoding='utf-8')
```

```python
production_missing_num.isna().sum()
```

```python
production_missing_num['thickness'].unique()
```

```python
# 欠損値の削除
production_missing_num.replace('None', np.nan, inplace=True)
production_missing_num.dropna(subset=['thickness'], inplace=True)
```

```python
# 欠損値の補完：平均値
production_missing_num.replace('None', np.nan, inplace=True)
production_missing_num['thickness'] = production_missing_num['thickness'].astype('float64')
thickness_mean = production_missing_num['thickness'].mean()
production_missing_num['thickness'].fillna(thickness_mean, inplace=True)
```

```python
# SimpleImputerを使った補完
from sklearn.impute import SimpleImputer

SimpleImputer().fit_transform(production_missing_num[['thickness']])
  # 引数に、strategy='most_frequent'を指定すると、最頻値で補完する
  # missing_values=Noneも可能
  # add_indicator=Trueで、欠損値の有無を示す列を追加できる。
```

```python
# 多変量補完
# 参考：https://qiita.com/FukuharaYohei/items/9830d5760595619352a5
# Scikit-learnのIterativeImputerやKNNImputerを使う
from sklearn.impute import IterativeImputer
from sklearn.experimental import enable_iterative_imputer
IterativeImputer().fit_transform(production_missing_num[['thickness', 'length']])

# ランダムフォレストを使った補完
from sklearn.ensemble import RandomForestRegressor
IterativeImputer(estimator=RandomForestRegressor()).fit_transform(production_missing_num[['thickness', 'length']])

# KNNを使った補完
from sklearn.impute import KNNImputer
KNNImputer(n_neighbors=2).fit_transform(production_missing_num[['thickness', 'length']])
```

```python
# カテゴリ型
# astype()の必要なし
customer_tb['sex_is_male'] = (customer_tb['sex']=='man').astype('bool')
customer_tb.head()

# カテゴリ型に変換
customer_tb['sex_c'] = customer_tb['sex'].astype('category')
# indexデータはcodesに格納されている
customer_tb['sex_c'].cat.codes
# マスタデータはcategoriesに格納されている
customer_tb['sex_c'].cat.categories
```

```python
# ダミー変数化
customer_tb['sex'] = pd.Categorical(customer_tb['sex'])
dummy_vars = pd.get_dummies(customer_tb['sex'], drop_first=True)
```

```python
# カテゴリ値の集約
customer_tb['age_rank'] = \
  pd.Categorical(np.floor(customer_tb['age']/10)*10)

customer_tb['age_rank'].cat.add_categories(['60以上'], inplace=True)
customer_tb.loc[customer_tb['age_rank'] \
  .isin([60.0, 70.0, 80.0]), 'age_rank'] = '60以上'
customer_tb['age_rank'].cat.remove_unused_categories(inplace=True)
```

```python
# カテゴリ値の組み合わせ
customer_tb['sex_and_age'] = pd.Categorical(
  customer_tb[['sex', 'age']]
  .apply(lambda x: '{}_{}'.format(x[0], np.floor(x[1]/10)*10), axis=1)
)
```

```python
# カテゴリ型の数値化
# 過学習を引き起こしたり、データの本来の意味を失ったりするので、注意が必要
# 例：カテゴリ値ごとに製造障害率を計算し、カテゴリ値に代用
production = pd.read_csv('awesomebook-master/data/production.csv', encoding='utf-8')

fault_cnt_per_type = production \
  .query('fault_fig') \
  .groupby('type') \
  .count()

type_cnt = production.groupby('type')['fault_fig'].count()
production['type_fault_rate'] = production[['type', 'fault_fig']] \
  .apply(lambda x: (fault_cnt_per_type[x[0]] - int(x[1])) / (type_cnt[x[0] - 1]), axis=1)
```
